## Business Scenario

The company sells products across multiple categories and regions. Management wants data-driven insights for next-quarter strategy.

The retail orders dataset contains:

- Order Date
- Product Category and Sub-Category
- Region
- Sales
- Profit
- Quantity

The main objective is to create Tableau visualizations and summarize insights that help management improve sales, profitability, and regional performance.

## Task 1: Category Performance Analysis

**Business Need:** Management wants to know which product categories generate the highest revenue.

**Recommended Chart:** Bar Chart

**Tableau Steps:**

1. Open Tableau and connect to the `Retail_Dataset`.
2. Drag **Category** to the **Rows** shelf.
3. Drag **Sales** to the **Columns** shelf.
4. Tableau will automatically apply **SUM(Sales)**.
5. Sort the bars in descending order.
6. Add data labels to clearly show total sales for each category.

**Answer / Insight:**  
The category with the highest total sales is the top-performing revenue category. A bar chart is suitable because it clearly compares sales values across different categories. Management can use this chart to identify which category should receive more focus in marketing, stock planning, and promotions.

## Task 2: Sales Trend Analysis

**Business Need:** Leadership wants to understand how sales fluctuate over time to plan promotions and inventory.

**Recommended Chart:** Line Chart

**Tableau Steps:**

1. Drag **Order Date** to the **Columns** shelf.
2. Change the date level to **Month**.
3. Drag **Sales** to the **Rows** shelf.
4. Ensure Sales is aggregated as **SUM(Sales)**.
5. Convert the chart type to **Line Chart**, if Tableau does not automatically select it.
6. Add labels or tooltips for monthly sales values.

**Answer / Insight:**  
Monthly sales trend analysis helps identify peak and slow periods. A line chart is ideal because it shows movement over time. Management can use the trend to plan inventory before high-sales months and create promotional campaigns during slow months.

## Task 3: Regional Contribution Analysis

**Business Need:** Regional managers want clarity on how much each region contributes to overall sales.

**Recommended Chart:** Pie Chart or Bar Chart

**Tableau Steps:**

1. Drag **Region** to the view.
2. Drag **Sales** to the view and aggregate it as **SUM(Sales)**.
3. Use a **Pie Chart** to show percentage contribution, or use a **Bar Chart** for easier comparison.
4. Add labels showing total sales and percentage contribution.
5. Sort regions from highest to lowest sales if using a bar chart.

**Answer / Insight:**  
The region with the highest total sales is the strongest contributor to business revenue. The lowest-performing region may require deeper analysis, better marketing, discount strategies, or supply-chain improvement.

## Task 4: Profitability Investigation

**Business Need:** Finance suspects that some high-sales orders may be reducing overall profit.

**Recommended Chart:** Scatter Plot

**Tableau Steps:**

1. Drag **Sales** to the **Columns** shelf.
2. Drag **Profit** to the **Rows** shelf.
3. Drag **Sub-Category** or **Product Name** to **Marks → Detail**.
4. Drag **Category** to **Color** for better segmentation.
5. Drag **Quantity** to **Size**, if product-level impact needs to be shown.
6. Add a reference line at Profit = 0 to separate profitable and loss-making orders.

**Answer / Insight:**  
A scatter plot helps identify products or orders with high sales but low or negative profit. These orders may have heavy discounts, high costs, or poor pricing. Management can review such products and improve pricing, reduce discounts, or discontinue low-profit items.

## Task 5: Business Growth Evaluation

**Business Need:** Senior management wants to assess whether the business is growing consistently.

**Recommended Chart:** Cumulative Line Chart / Running Total Line Chart

**Tableau Steps:**

1. Drag **Order Date** to the **Columns** shelf.
2. Change the date level to **Month**.
3. Drag **Sales** to the **Rows** shelf.
4. Right-click on **SUM(Sales)** and select **Quick Table Calculation → Running Total**.
5. Format the line chart clearly with title and labels.

**Answer / Insight:**  
Cumulative sales growth shows whether total business revenue is increasing steadily over time. If the running total line rises consistently, the business is growing. If the line becomes flat, it indicates weak growth and management should investigate sales slowdown.

## Task 6: Category vs Region Performance

**Business Need:** Operations wants to know which regions perform well for each product category.

**Recommended Chart:** Highlight Table / Heat Map

**Tableau Steps:**

1. Drag **Category** to the **Rows** shelf.
2. Drag **Region** to the **Columns** shelf.
3. Drag **Sales** to **Color** on the Marks card.
4. Drag **Sales** again to **Label**.
5. Select **Square** mark type to create a heat map.
6. Use darker color intensity to represent higher sales.

**Answer / Insight:**  
A heat map helps compare category-wise sales across regions. Strong areas show high sales for a specific category-region combination, while weak areas show low contribution. Operations can use this to decide region-wise inventory distribution and category-specific campaigns.

## Task 7: Product Prioritization

**Business Need:** Product team wants to focus on products that create maximum business impact.

**Recommended Chart:** Bubble Chart or Scatter Plot

**Tableau Steps:**

1. Drag **Sales** to the **Columns** shelf.
2. Drag **Profit** to the **Rows** shelf.
3. Drag **Product Name** or **Sub-Category** to **Marks → Detail**.
4. Drag **Quantity** to **Size**.
5. Drag **Category** to **Color**.
6. Analyze products with high sales, high profit, and high quantity.

**Answer / Insight:**  
Priority products are those that generate strong sales, strong profit, and high quantity movement. Products with high sales but low profit should be reviewed. Products with high profit and strong demand should receive more marketing and inventory support.

## Task 8: Executive Dashboard Creation

**Business Need:** CEO wants a single-view dashboard that summarizes key insights.

**Dashboard Requirement:** Combine at least three visualizations and apply filters such as Region or Category.

**Recommended Dashboard Components:**

1. **Total Sales by Category** — Bar Chart
2. **Monthly Sales Trend** — Line Chart
3. **Region-wise Sales Contribution** — Bar Chart or Pie Chart
4. **Sales vs Profit Analysis** — Scatter Plot
5. Optional KPI Cards: Total Sales, Total Profit, Total Quantity

**Tableau Steps:**

1. Create each required worksheet separately.
2. Click **New Dashboard**.
3. Drag at least three worksheets into the dashboard.
4. Add filters for **Region** and **Category**.
5. Apply filters to all relevant worksheets using **Apply to Worksheets → All Using This Data Source**.
6. Add a clear title such as **Retail Sales Performance Dashboard**.
7. Arrange charts neatly for executive-level readability.

**Answer / Insight:**  
An executive dashboard gives management a single view of revenue, trends, profitability, and regional performance. It saves time and helps decision-makers quickly identify areas that need action.

## Q9: Chart Selection Justification

Different charts were selected because each business question requires a different type of analysis.

| Business Question | Chart Used | Reason |
|---|---|---|
| Which category has the highest sales? | Bar Chart | Best for comparing values across categories. |
| How do sales change over time? | Line Chart | Best for showing trends over months or dates. |
| Which region contributes the most sales? | Pie Chart / Bar Chart | Shows contribution and comparison by region. |
| Which orders are low-profit or loss-making? | Scatter Plot | Shows the relationship between Sales and Profit. |
| Is business growing over time? | Running Total Line Chart | Shows cumulative business growth. |
| Which category performs best in each region? | Heat Map | Shows strong and weak performance areas using color intensity. |
| Which products should be prioritized? | Bubble Chart / Scatter Plot | Combines Sales, Profit, and Quantity in one view. |

**Conclusion:**  
Chart selection depends on the business problem. Comparison questions need bar charts, time-based questions need line charts, relationship-based questions need scatter plots, and matrix-based performance questions need heat maps.

## Q1. Why is publishing & sharing more important than building dashboards?
- Enables collaboration across teams
- Provides real-time access to stakeholders
- Ensures decision-making using updated data
- Makes reports accessible on web & mobile
- Transforms insights into business value


## Q2. Power BI Service vs Desktop
**Power BI Desktop:** Used for data modeling, report creation

**Power BI Service:** Cloud platform used for:
- Publishing reports
- Sharing dashboards
- Collaboration
- Scheduled refresh

**Role:** Desktop = Development | Service = Distribution


## Q3. What is Publishing?
Publishing means uploading a report from Desktop to Power BI Service.

**Outcomes:**
1. Report becomes accessible online
2. Can be shared with users


## Q4. Steps to Publish Report
1. Open report in Power BI Desktop
2. Click 'Publish'
3. Sign in to Power BI account
4. Select workspace
5. Confirm upload
6. View in Power BI Service


## Q5. Report vs Dashboard
| Feature | Report | Dashboard |
|--------|--------|-----------|
| Pages | Multiple | Single |
| Source | Dataset | Multiple reports |
| Usage | Detailed analysis | Summary view |
| Interactivity | High | Limited |


## Q6. Sharing Methods
**1. Direct Sharing**
- Share report via email
- Use case: Small team

**2. Workspace Access**
- Users added with roles
- Use case: Team collaboration

**3. Power BI Apps**
- Packaged reports
- Use case: Organization-wide distribution


## Q7. Workspace Roles
- Admin: Full control
- Member: Edit & share
- Contributor: Edit only
- Viewer: Read-only

Access control ensures security based on role permissions.


## Q8. Why Apps are more secure?
- Centralized distribution
- Controlled updates
- Version management
- Restricted access
- Better governance


## Q9. Row-Level Security (RLS)
RLS restricts data visibility based on user roles.

Example:
- Sales Manager sees only their region data
- Uses filters & roles


## Q10. Risks without proper sharing methods
1. Data leakage
2. Unauthorized access
3. Misinterpretation of data

**Prevention:**
- Use Apps
- Apply workspace roles
- Implement RLS

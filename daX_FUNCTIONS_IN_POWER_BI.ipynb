## Task 1: Create a calculated column TotalPrice using Quantity and UnitPrice

TotalPrice = 'Amazon Store Sales Dataset'[Quantity] * 'Amazon Store Sales Dataset'[UnitPrice]

Explanation

This calculated column multiplies the quantity sold by the unit price of each product to get the total price before discount.

## Task 2: Create a calculated column DiscountedPrice

DiscountedPrice =
'Amazon Store Sales Dataset'[TotalPrice] * (1 - 'Amazon Store Sales Dataset'[Discount] / 100)


Explanation

This column calculates the final price after applying the discount percentage.

## Task 3: Create a measure Average Order Value (AOV)

Average Order Value (AOV) =
DIVIDE(
    SUM('Amazon Store Sales Dataset'[SalesAmount]),
    DISTINCTCOUNT('Amazon Store Sales Dataset'[OrderID])
)


Explanation

This measure calculates the average sales value per order by dividing total sales amount by the number of unique orders.




## Task 4: Create a measure for cumulative Year-to-Date (YTD) Sales based on OrderDate

YTD Sales =
TOTALYTD(
    SUM('Amazon Store Sales Dataset'[SalesAmount]),
    'Amazon Store Sales Dataset'[OrderDate]
)

Explanation

This measure returns cumulative sales from the start of the year up to the current date context.

## Task 5: Create a measure to rank each product according to Total Sales

Product Rank =
RANKX(
    ALL('Amazon Store Sales Dataset'[ProductID]),
    CALCULATE(SUM('Amazon Store Sales Dataset'[SalesAmount])),
    ,
    DESC,
    DENSE
)

Explanation

This measure ranks all products based on total sales in descending order.
The product with the highest sales gets rank 1.

## Task 6: Create a calculated column classifying orders as High Value or Low Value

Order Value Category =
IF(
    'Amazon Store Sales Dataset'[SalesAmount] > 2000,
    "High Value",
    "Low Value"
)

Explanation

This column checks each order’s sales amount:

Greater than 2000 → High Value
Otherwise → Low Value


## Task 7: Create a measure to calculate total sales from the Online channel only

Online Sales =
CALCULATE(
    SUM('Amazon Store Sales Dataset'[SalesAmount]),
    'Amazon Store Sales Dataset'[SalesChannel] = "Online"
)



Explanation

This measure filters the dataset for only Online sales and then sums the sales amount.

## Task 8: Create a measure called Date Difference using DATEDIFF\

Date Difference =
DATEDIFF(
    MAX('Amazon Store Sales Dataset'[OrderDate]),
    TODAY(),
    DAY
)

Explanation

This measure calculates the number of days between the latest order date in the dataset and today’s date.

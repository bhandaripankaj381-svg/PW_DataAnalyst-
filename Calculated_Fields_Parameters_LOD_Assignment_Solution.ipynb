# Calculated Fields, Parameters & LOD Expressions — Solution

## Theoretical Questions

### Question 1: What is a Calculated Field and why is it important in business dashboards?

A **Calculated Field** in Tableau is a user-created field made by applying formulas, conditions, functions, or logical expressions on existing dataset fields.

Calculated fields are important in business dashboards because they help create new business metrics that may not already exist in the raw data. For example, profit margin, delivery status, loss flag, sales category, and performance level can be created using calculated fields.

**Example:**

```tableau
[Profit] / [Sales]
```

This can be used to calculate **Profit Margin** and help management understand profitability more clearly.

### Question 2: Why can’t parameters change visuals directly in Tableau?

Parameters cannot change visuals directly because a parameter is only a **single user-controlled value**. It does not automatically interact with the data or chart by itself.

To make a parameter affect a visual, it must be connected with a **Calculated Field**, **Filter**, or **Reference Line**.

**Example:** If we create a parameter named **Top N**, it will not automatically show top products. We need a calculated field or filter logic that uses that parameter to display only the selected number of products.

### Question 3: When should EXCLUDE LOD be used?

**EXCLUDE LOD** should be used when we want to remove a specific dimension from the level of detail of a view while performing aggregation.

It is useful when the chart contains detailed dimensions but the analysis needs a higher-level result.

**Example:** If a view shows Sales by **Category** and **Sub-Category**, but we want to calculate sales only at the **Category** level, we can exclude Sub-Category.

```tableau
{ EXCLUDE [Sub-Category] : SUM([Sales]) }
```

This removes Sub-Category from the calculation and gives a broader category-level result.

### Question 4: Difference between FIXED and INCLUDE LOD expressions

| Basis | FIXED LOD | INCLUDE LOD |
|---|---|---|
| Meaning | Calculates value at a fixed dimension level | Adds extra dimension detail into the calculation |
| Dependency on View | Mostly independent of the dimensions shown in the view | Depends on the view and adds more detail to it |
| Use Case | City-level average sales regardless of category view | Product-level average sales while viewing Category-level analysis |
| Example | `{ FIXED [City] : AVG([Sales]) }` | `{ INCLUDE [Product_Name] : AVG([Sales]) }` |

**FIXED** is used when the calculation must stay fixed at a specific dimension level. **INCLUDE** is used when we want to calculate at a more detailed level than what is currently shown in the view.

## Practical Questions

### Task 1: Delivery Performance Classification — Calculated Field

**Business Need:** Operations wants to track delivery efficiency.

**Calculated Field Name:** `Delivery Status`

**Formula:**

```tableau
IF [Delivery_Time_Min] <= 30 THEN "Fast"
ELSE "Delayed"
END
```

**Explanation:**

This calculated field classifies every order based on delivery time. If the delivery time is 30 minutes or less, the order is marked as **Fast**. Otherwise, it is marked as **Delayed**.

**Suggested Visualization:**

Create a bar chart showing count of orders by `Delivery Status`.

**Steps:**

1. Create calculated field named `Delivery Status`.
2. Drag `Delivery Status` to Columns.
3. Drag `Order_ID` to Rows.
4. Change aggregation of `Order_ID` to Count.
5. Use color to separate Fast and Delayed orders.

**Insight:**

This chart helps identify whether Blinkit is meeting its quick delivery promise or whether delayed orders are increasing.

### Task 2: Loss-Making Orders Identification

**Business Need:** High sales do not always mean high profit.

**Calculated Field Name:** `Loss Status`

**Formula:**

```tableau
IF [Profit] < 0 THEN "Loss-Making Order"
ELSE "Profitable Order"
END
```

**Alternative Calculated Field Name:** `Loss Flag`

```tableau
IF [Profit] < 0 THEN 1
ELSE 0
END
```

**Suggested Visualization:**

Create a scatter plot using `Sales` and `Profit`.

**Steps:**

1. Drag `Sales` to Columns.
2. Drag `Profit` to Rows.
3. Drag `Order_ID` to Detail.
4. Drag `Loss Status` to Color.
5. Analyze points below zero profit.

**Insight:**

Orders with high sales but negative profit indicate pricing issues, high cost, discount problems, or operational inefficiency. Management should review these orders to reduce losses.

### Task 3: Top N Products Analysis — Parameter

**Business Need:** The product team wants to focus on top-performing products.

**Parameter Name:** `Top N Products`

**Parameter Settings:**

| Setting | Value |
|---|---|
| Data Type | Integer |
| Current Value | 10 |
| Allowable Values | Range |
| Minimum | 5 |
| Maximum | 20 |
| Step Size | 1 |

**Calculated Field Name:** `Top N Product Filter`

**Formula:**

```tableau
RANK(SUM([Sales])) <= [Top N Products]
```

**Steps:**

1. Create a parameter named `Top N Products`.
2. Create calculated field `Top N Product Filter` using the above formula.
3. Drag `Product_Name` to Rows.
4. Drag `Sales` to Columns.
5. Sort products in descending order by Sales.
6. Drag `Top N Product Filter` to Filters.
7. Select `True`.
8. Show parameter control so the user can dynamically change Top N value.

**Insight:**

This analysis helps the product team identify the products generating maximum sales and prioritize them for inventory, marketing, and promotion planning.

### Task 4: City-Level Benchmarking — FIXED LOD

**Business Need:** Compare city performance independent of category filters.

**Calculated Field Name:** `City Average Sales`

**Formula:**

```tableau
{ FIXED [City] : AVG([Sales]) }
```

**Explanation:**

This calculation fixes the average sales calculation at the City level. It gives the average sales for each city even when the view contains other dimensions.

**Suggested Visualization:**

Create a bar chart showing `City` vs `City Average Sales`.

**Steps:**

1. Create calculated field `City Average Sales`.
2. Drag `City` to Rows.
3. Drag `City Average Sales` to Columns.
4. Sort cities in descending order.
5. Add `Category` as filter if needed.
6. Analyze city-level performance.

**Insight:**

This helps management compare cities fairly and identify cities with strong or weak average sales performance.

### Task 5: Customer/Product Detail Impact — INCLUDE LOD

**Business Need:** The sales team wants average sales considering product-level detail, even when product is not shown.

**Calculated Field Name:** `Product Level Average Sales`

**Formula:**

```tableau
{ INCLUDE [Product_Name] : AVG([Sales]) }
```

**Explanation:**

This calculation includes product-level detail in the average sales calculation even if `Product_Name` is not present in the view. The result can be analyzed at a higher level such as Category or City.

**Suggested Visualization:**

Create a bar chart showing `Category` vs `Product Level Average Sales`.

**Steps:**

1. Create calculated field `Product Level Average Sales`.
2. Drag `Category` to Rows.
3. Drag `Product Level Average Sales` to Columns.
4. Add `City` or `Payment_Mode` as filter if needed.
5. Compare product-level average sales across categories.

**Insight:**

This helps identify which category performs better after considering product-level detail. It provides a more accurate view than simply calculating average sales directly at the category level.

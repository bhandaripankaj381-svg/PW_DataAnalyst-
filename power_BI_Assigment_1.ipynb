##Question 1: What is Power BI and why is it used in businesses?

Anser : Power BI is a Business Intelligence (BI) tool developed by Microsoft used to analyze data and create interactive dashboards and reports.

Why businesses use Power BI:

Data visualization (charts, graphs)
Decision making
Data analysis from multiple sources
Real time reporting
Finding trends and patterns

## Question 2: Name and explain the three main components of Power BI ?

Anser :
 1. Power BI Desktop
 - Main software used to create reports and dashboards
Data cleaning and modeling happens here

2. Power BI Service (Cloud)
- Online platform
Used to share reports and dashboards
3. Power BI Mobile
- Mobile app
Used to view reports on phone

## Question 3: Explain the Power BI workflow
Anser : Power BI workflow steps:

Connect Data → Import data from Excel, SQL etc.

Transform Data → Clean data in Power Query

Data Modeling → Create relationships between tables

Create Visuals → Charts, tables, dashboards

Publish Report → Upload to Power BI Service

Share Insights → Share with team

## Question 6: Define Data View, Report View, and Model View

Data View
Shows table data in rows and columns
Used to inspect data

Report View
Used to create charts and dashboards
Main visualization area

Model View
Shows relationships between tables
Used for data modeling

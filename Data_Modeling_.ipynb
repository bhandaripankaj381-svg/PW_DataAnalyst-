## Question 1: What is Data Loading in Power BI and why is it considered the first step of analysis?

Ans Data loading in Power BI means importing data from different sources like Excel, SQL databases, CSV files, or web sources into Power BI for analysis.

It is considered the first step because:

Analysis cannot start without data
Data must be available before cleaning and modeling
All transformations depend on loaded data

## Question 2: Explain the difference between "Load" and "Transform Data" in Power BI.
Ans
| Load                                | Transform Data                    |
| ----------------------------------- | --------------------------------- |
| Directly imports data into Power BI | Opens Power Query Editor          |
| No changes made to data             | Used for data cleaning            |
| Faster process                      | Used for preprocessing            |
| Used when data is already clean     | Used when data needs modification |


## Question 3: What is a Fact Table and a Dimension Table? Give examples.

Ans :
Fact Table:

A fact table contains numerical data (measurable values) and foreign keys.

Examples:

Sales Amount
Property Price
Rent Amount
Transaction Value

Example from real estate dataset:

Property price
Area
Sale amount
Dimension Table:

A dimension table contains descriptive information about facts.

Examples:

Owner Name
City
Property Type
Location

| Fact Table     | Dimension Table   |
| -------------- | ----------------- |
| Numerical data | Descriptive data  |
| Measures       | Attributes        |
| Example: Sales | Example: Customer |


## Question 4: Why is Star Schema preferred over Snowflake Schema in Power BI?

Answer:
Star schema is preferred because:

Simple structure
Faster query performance
Easy to understand
Better for reporting

| Star Schema        | Snowflake Schema   |
| ------------------ | ------------------ |
| Simple structure   | Complex structure  |
| Less joins         | More joins         |
| Faster performance | Slower performance |
| Easy to use        | Hard to manage     |


## Question 5: Identify and remove duplicate records based on Date, Country, and State

Ans
Go to Transform Data (Power Query)
Select columns:
Date
Country
State
Go to Home tab
Click Remove Rows
Click Remove Duplicates

Concept:
Power BI selected columns ke basis par duplicate rows remove kar dega.

## Question 6: Identify and replace null values in Vaccination_Status

Ans
Steps:

Open Power Query
Select Vaccination_Status column
Click Transform tab
Click Replace Values
Replace:
Value to find → null
Replace with → Not Vaccinated (or Unknown)

Alternative method:
Transform → Replace Values → null → "No Data"

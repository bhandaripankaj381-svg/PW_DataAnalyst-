## Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.

Tableau is a Business Intelligence (BI) and data visualization tool used to connect, analyze, and present data through interactive charts, dashboards, and reports. It helps users convert raw data into meaningful business insights without writing complex code.

Tableau is preferred over traditional spreadsheet tools because:

1. **Interactive dashboards:** Tableau allows users to create dynamic dashboards with filters, drill-downs, and interactive views.
2. **Better data visualization:** It provides advanced charts, maps, and visual analytics that are easier to understand than large spreadsheet tables.
3. **Handles large datasets efficiently:** Tableau can connect with large datasets and databases more effectively than normal spreadsheets.
4. **Easy data connection:** Tableau can connect to Excel, CSV, databases, cloud platforms, and live data sources.
5. **Faster decision-making:** Users can quickly identify trends, patterns, and outliers through visual analysis.

### Industries where Tableau is commonly used

**1. Retail Industry**  
Use case: A retail company can use Tableau to analyze sales, profit, product categories, customer segments, and regional performance. For example, managers can identify which product category generates the highest profit and which region needs improvement.

**2. Healthcare Industry**  
Use case: Hospitals can use Tableau to analyze patient visits, treatment costs, hospital performance, and resource utilization. This helps in improving service quality and operational planning.

### a) Main types of fields available

In Tableau's Sample – Superstore dataset, the main types of fields available are:

1. **Time-based fields:** These fields are related to dates and time.  
   Examples: Order Date, Ship Date

## Q3. Create a Bar Chart showing Total Sales by Category

To create a bar chart showing **Total Sales by Category** in Tableau:

1. Connect to the Sample – Superstore dataset.
2. Drag **Category** to the **Rows** shelf.
3. Drag **Sales** to the **Columns** shelf.
4. Tableau automatically creates a bar chart.
5. Make sure the Marks type is set to **Bar**.

### Explanation

| Requirement | Answer |
|---|---|
| Field placed on Rows | Category |
| Field placed on Columns | Sales |
| Aggregation applied to Sales | SUM(Sales) |

Tableau applies **SUM** aggregation by default to the Sales measure. This means the chart displays total sales for each product category such as Furniture, Office Supplies, and Technology.


## Q4. Insight from comparing Sales and Profit together

Comparing **Sales** and **Profit** helps businesses understand whether high sales are actually generating good profit. Sometimes a product may have high sales but low profit because of high discounts, high cost, or poor margins.

For example:

- High Sales + High Profit = Strong performing product/category
- High Sales + Low Profit = Revenue is good, but margin needs improvement
- Low Sales + High Profit = Niche product with good profitability
- Low Sales + Low Profit = Weak performing product/category

### Chart to analyze Profit by Sub-Category

To create a chart for **Profit by Sub-Category**:

1. Drag **Sub-Category** to the **Rows** shelf.
2. Drag **Profit** to the **Columns** shelf.
3. Tableau will create a horizontal bar chart.
4. Sort the chart in descending order by Profit.
5. Use color to highlight positive and negative profit values.

This chart helps identify which sub-categories are profitable and which are causing losses.


## Q5. Create a Line Chart to show Sales Trend over Order Date

To create a line chart showing **Sales Trend over Order Date**:

1. Drag **Order Date** to the **Columns** shelf.
2. Drag **Sales** to the **Rows** shelf.
3. Change the Marks type to **Line** if Tableau does not automatically select it.
4. Change the date level from Year to **Month** for detailed trend analysis.

### Which date level would I choose?

I would choose **Month** because monthly analysis gives a more detailed view of sales trends compared to yearly analysis. It helps identify seasonal patterns, monthly growth, and sales fluctuations.

### Why are line charts suitable for time-based analysis?

Line charts are suitable for time-based analysis because they show how values change over a continuous period. They make it easy to identify trends, increases, decreases, seasonality, and patterns over time.


## Q6. Apply a Basic Filter for Region = West and Category = Technology

To display data only for **Region = West** and **Category = Technology**:

1. Drag **Region** to the **Filters** shelf.
2. In the filter window, select only **West**.
3. Click **OK**.
4. Drag **Category** to the **Filters** shelf.
5. In the filter window, select only **Technology**.
6. Click **OK**.

### Shelf used

The shelf used for applying filters is the **Filters shelf**.

After applying these filters, Tableau will show only those records where the region is West and the category is Technology.

## Q8. Difference between Worksheet and Dashboard in Tableau

| Worksheet | Dashboard |
|---|---|
| A worksheet is a single view or chart created from data. | A dashboard is a collection of multiple worksheets/views in one place. |
| It focuses on one specific analysis. | It combines multiple analyses for a complete business view. |
| Example: Sales by Category bar chart. | Example: Sales dashboard with bar chart, line chart, map, and filters. |

### Why is a dashboard more useful for decision-makers?

A dashboard is more useful for decision-makers because it provides multiple insights in one place. Managers can view KPIs, trends, comparisons, maps, and filters together. This helps them make faster and better decisions without checking separate reports one by one.


## Q9. What is Tableau Prep, and why is it important in real-world data projects?

Tableau Prep is a data preparation tool used to clean, combine, reshape, and prepare data before analysis in Tableau. It helps users create clean and structured datasets for accurate reporting and visualization.

Tableau Prep is important in real-world data projects because raw data often contains errors such as null values, duplicate records, incorrect data types, spelling mistakes, and inconsistent formats. Clean data improves the accuracy and reliability of dashboards.

### Steps to remove null values

1. Open the dataset in Tableau Prep.
2. Add a cleaning step.
3. Select the field that contains null values.
4. Filter out null values or replace them with a suitable value.
5. Review the cleaned data.

### Steps to remove duplicate records

1. Open the dataset in Tableau Prep.
2. Add a cleaning step.
3. Select the fields that define a duplicate record.
4. Use grouping or remove duplicate logic.
5. Keep only unique records in the output.

### Steps to change incorrect data types

1. Open the dataset in Tableau Prep.
2. Go to the field with the incorrect data type.
3. Click the data type icon beside the field name.
4. Select the correct data type such as String, Number, Date, or Boolean.
5. Verify that the values are correctly converted.

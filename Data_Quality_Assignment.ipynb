## Q1: Define Data Quality in ETL. Why more than cleaning?

 Ans : Data Quality refers to the accuracy, completeness, consistency, validity, and reliability of data in an ETL pipeline.

It is more than just data cleaning because:

Cleaning only removes errors
Data quality ensures:
Correct format (validation)
Consistency across systems
Business rule compliance
Reliability for decision-making

👉 So, it covers the entire lifecycle, not just fixing errors.

## Q2: Why poor data quality leads to wrong decisions?

Ans : Poor data quality causes:

Incorrect KPIs (e.g., wrong revenue)
Duplicate records → inflated numbers
Missing values → incomplete analysis
Inconsistent data → confusion in dashboards

👉 Result:

Misleading dashboards
Wrong business strategies
Financial losses

# Q3: What is duplicate data? Causes in ETL ?

Ans : Duplicate data means same record appearing multiple times.

Causes:
Multiple data sources (same data coming from different systems)
Improper joins/merges in ETL
Reprocessing of data (same file loaded again)

## Q4: Types of duplicates

Ans :
## 1 Exact Duplicate
Completely same record
Example: Same Customer_ID, Product_ID, Date, Amount
## 2 Partial Duplicate
ome fields same, some different
Example: Same customer, different amount
## 3 Fuzzy Duplicate
Slight variations (spelling, format)
Example: “Rahul Mehta” vs “Rahul M.”

## Q5: Why validate during transformation?

Ans : Validation during transformation is important because:

Errors are caught early
Prevents bad data from entering database
Reduces rework
Improves pipeline efficiency

👉 After loading, fixing becomes costly and complex.



## Q6: Business rules in validation ?

Ans : Business rules define conditions data must satisfy.

Example:
Txn_Amount should not be NULL
Quantity must be > 0

👉 If violated → data is invalid

## Q7: SQL query for duplicate keys
Ans :SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS duplicate_count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1;

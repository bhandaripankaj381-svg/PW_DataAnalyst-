# Employee_Productivity_Dataset Assignment


```
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

df=pd.read_csv('/mnt/data/6c4df4c5-8d93-4271-8275-a2215f63c2bd.csv')

# Q1
df['Age']=df['Age'].fillna(df['Age'].median())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week']=df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score']=df['Performance_Score'].fillna(df['Performance_Score'].mean())
print("Q1");display(df.head())

# Q2
le=LabelEncoder()
for c in ['Gender','Department']:
    df[c]=le.fit_transform(df[c])
print("Q2");display(df[['Gender','Department']].head())

# Q3
df=pd.get_dummies(df,columns=['Work_Mode','Location'])
print("Q3",df.shape);display(df.head())

# Q4
mm=MinMaxScaler()
df[['Salary','Hours_Worked_Per_Week']]=mm.fit_transform(df[['Salary','Hours_Worked_Per_Week']])
print("Q4");display(df[['Salary','Hours_Worked_Per_Week']].head())

# Q5
ss=StandardScaler()
df[['Age','Projects_Completed']]=ss.fit_transform(df[['Age','Projects_Completed']])
print("Q5");display(df[['Age','Projects_Completed']].head())

# Q6
tmp=pd.read_csv('/mnt/data/6c4df4c5-8d93-4271-8275-a2215f63c2bd.csv')
comp=pd.DataFrame()
comp['MinMax']=MinMaxScaler().fit_transform(tmp[['Salary']]).ravel()
comp['Standard']=StandardScaler().fit_transform(tmp[['Salary']].fillna(tmp['Salary'].mean())).ravel()
print("Q6");display(comp.head())

# Q7
orig=pd.read_csv('/mnt/data/6c4df4c5-8d93-4271-8275-a2215f63c2bd.csv')
num=['Age','Salary','Hours_Worked_Per_Week','Projects_Completed','Performance_Score']
cat=['Gender','Department','Work_Mode','Location']
pre=ColumnTransformer([
('num',StandardScaler(),num),
('cat',OneHotEncoder(handle_unknown='ignore'),cat)])
pipe=Pipeline([('preprocessor',pre)])
print("Q7");print(pipe)

# Q8
X=pipe.fit_transform(orig)
print("Q8 shape",X.shape)

# Q9
print("Q9: Scaling ensures features are on similar ranges, improving model performance and convergence.")

# Q10
print("Q10: ML algorithms require numeric input, so categorical values are encoded into numbers.")
```

# Titanic EDA Assignment

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/mnt/data/1a8a817f-d90e-4c9f-a6c2-97ec080b1dcb.csv')
df.head()
```

## Q1. Create a countplot to show the number of passengers by gender.

```
plt.figure(figsize=(5,4));sns.countplot(x='Sex',data=df);plt.show()
```

## Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.

```
df['Pclass'].value_counts().sort_index().plot(kind='bar',color=['red','green','blue']);plt.show()
```

## Q3. Create a histogram to show the distribution of passenger age.

```
sns.histplot(df['Age'].dropna(),kde=True);plt.show()
```

## Q4. Create a boxplot to detect outliers in the fare column.

```
sns.boxplot(y=df['Fare']);plt.show()
```

## Q5a. Show relationship between age and fare.

```
sns.scatterplot(x='Age',y='Fare',data=df);plt.show()
```

## Q5b. Analyze family size and fare.

```
df['FamilySize']=df['SibSp']+df['Parch'];sns.scatterplot(x='FamilySize',y='Fare',data=df);plt.show()
```

## Q6. Survival count by gender.

```
sns.countplot(x='Sex',hue='Survived',data=df);plt.show()
```

## Q7. Survival by passenger class.

```
sns.barplot(x='Pclass',y='Survived',data=df);plt.show()
```

## Q8. Correlation heatmap.

```
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True);plt.show()
```

## Q9. Highest average fare class.

```
print(df.groupby('Pclass')['Fare'].mean())
```

## Q10. Insights.

```
print('Insights: First class has highest fare. Fare has outliers. Survival varies by class and gender. Age mostly young adults. Family size weakly related to fare.')
```
